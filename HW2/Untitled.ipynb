```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
u = np.random.uniform(size=N)
v = np.random.uniform(size=N)
```

```python
x = 1 - np.sqrt(1 - u**2)
y = 1 - np.exp(-v)
```

```python
plt.scatter(x,y)
plt.xlim(0,1)
plt.ylim(0,1)

X = np.linspace(0,1)
plt.plot(X,1-X)

plt.show()
```
